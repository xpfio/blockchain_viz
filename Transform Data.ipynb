```python
import json

nodes = []
links = []

with open('data/address1.json') as data_file:    
    data = json.load(data_file)
#     data['txs']
    for transaction in data['txs']:
        inputs = transaction['inputs']
        outputs = transaction['out']
        for sender in inputs:
            for receiver in outputs:
#                 print receiver
                receiver_address = ''
                if 'addr' in receiver:
                    receiver_address = receiver['addr']
                links.append({
                    'source':sender['prev_out']['addr'],
                    'target':receiver_address,
                    'value':2
                })
                if sender['prev_out']['addr'] not in nodes:
                    nodes.append(sender['prev_out']['addr'])
                if receiver_address not in nodes:
                    nodes.append(receiver_address)
    
```

```python
list_nodes = []
for node in nodes:
    list_nodes.append({
        'id':node,
        'group':1
    })
```

```python
output = {
    'nodes':list_nodes,
    'links':links
}
with open('visualization/data_bitcoin.json', 'w') as outfile:
    json.dump(output, outfile)
```
